```python
import os
import pandas as pd
from openpyxl import load_workbook
```

```python
# Directory containing Excel files
directory = "/Users/shivamsoni/Downloads/Minaki SKU Data Generation"  # Replace with the actual directory path
```

```python
# Initialize an empty master DataFrame
master_df = pd.DataFrame()



# Function to extract a table starting from a specific row and stop at first blank row
def read_xlsx_table(file_path, table_start_index=6):

    if 'June 2024.xlsx' in file_path:
        table_start_index = 0

    # Read the Excel file with the given header row
    df = pd.read_excel(file_path, header=table_start_index)


    # update price($) and SKU
        
    if 'Unnamed: 1' in df.columns:
        df.rename(columns={'No.' : 'No.(1)', 'Unnamed: 1' : 'No.'},inplace=True)

    if 'Price($)' in df.columns:
        df.rename(columns={'Price($)' : 'Price()'},inplace=True)

    if 'SKU' in df.columns:
        df.rename(columns={'SKU' : 'No.'},inplace=True)

    if 'No.' and 'sku' in df.columns:
        df.rename(columns={'No.' : 'No.(1)', 'sku' : 'No.'},inplace=True)

    if 'April 2024.xlsx' in file_path:
        print(df.columns)    
    
    # Find the first row where all values are NaN
    first_blank_row = df.loc[df.isnull().all(axis=1)].index.min()
    
    # Strip all rows including and after the first blank row
    if pd.notna(first_blank_row):
        df = df.iloc[:first_blank_row]
        cols = ['No.','NO.', 'Color', 'Size', 'Price()', 'Quantity', 'Remark']
        df = df[cols]
    
    return df
```

```python
# Loop through all files in the directory
for file in os.listdir(directory):
    
    # Check if the file is an Excel file
    if file.endswith((".xlsx", ".xls")) and not file.startswith("~$"):
        file_path = os.path.join(directory, file)
        print(f"Reading file: {file_path}")
        
        # Extract table from the Excel file
        try:
            temp_df = read_xlsx_table(file_path)
            
            if not temp_df.empty:
                # Append to the master DataFrame
                master_df = pd.concat([master_df, temp_df], ignore_index=True)
        except Exception as e:
            print(f"Error reading {file}: {e}")
```

```python
# Show the master DataFrame
def display_master_df():
    print("\nFinal Master DataFrame:")
    print(master_df.head())  # Display the first few rows
```

```python
# Save the master DataFrame to a new Excel file
output_file = "master_output.xlsx"
master_df.to_excel(output_file, index=False)
print(f"Master DataFrame saved to {output_file}")
```

```python
# Display result
display_master_df()
```

```python
import bcrypt
import os
import pandas as pd
from dotenv import load_dotenv

# Load environment variables from .env
load_dotenv()

from sqlalchemy import create_engine

# Database connection (replace with your actual connection string)
POSTGRES_URI = os.getenv("POSTGRES_URI")
engine = create_engine(POSTGRES_URI)

# Users and their plaintext passwords
users = [
    {"username": "monika", "password": "123456", "email": "monika@example.com"},
    {"username": "bhavvam", "password": "123456", "email": "bhavvam@example.com"}
]

# Hash passwords for each user
for user in users:
    user['hashed_password'] = bcrypt.hashpw(user['password'].encode('utf-8'), bcrypt.gensalt()).decode('utf-8')

# Remove plaintext passwords to avoid storing them
for user in users:
    del user['password']

# Convert users list to a Pandas DataFrame
df = pd.DataFrame(users)

# Insert data into the 'users' table
try:
    df.to_sql("users", engine, if_exists="append", index=False)
    print("Data inserted successfully!")
except Exception as e:
    print(f"Error inserting data: {e}")
```
